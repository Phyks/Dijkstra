```python
%matplotlib inline
import matplotlib.pyplot as plt
import matplotlib
import numpy as np
```

```python
vertice = []
edge = []
time = []
f = open('test_out_fib', 'r')

for line in f.readlines()[1:]:
    v, e, t = line.split()
    vertice.append(float(v))
    edge.append(float(e))
    time.append(float(t))
```

```python
plt.figure(figsize=(5, 4), dpi=80)
plt.plot(edge,time, ".")
plt.savefig('report/edge.pdf')
```

```python
plt.figure(figsize=(5, 4), dpi=80)
plt.plot(vertice, time, ".")
plt.savefig('report/vertice.pdf')
```

```python
plt.figure(figsize=(5, 4), dpi=400)
corrected = [(e + v*np.log10(v))/1000 for e, v in zip(edge, vertice) ]
plt.plot(corrected, time, ".")
plt.ylabel("Temps (s)")
plt.xlabel("$|E| + |V| \log |V|$ ($1/1000$)")
plt.plot([0, max(corrected)], [0,2.2])
plt.savefig('report/reg.pdf')
```

```python
plt.figure(figsize=(5, 4), dpi=400)
corrected = [v*v for e, v in zip(edge, vertice) ]
plt.plot(corrected, time, ".")
plt.plot([0.0, max(corrected)], [0.03, 0.80])
plt.ylabel("Temps (s)")
plt.xlabel("$|V|^2$")
```
